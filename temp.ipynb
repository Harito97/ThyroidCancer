```python
import numpy as np

# Confusion matrix
conf_matrix = np.array([[1787, 980, 614], [960, 2279, 142], [634, 122, 2625]])

# True Positives (TP), True Negatives (TN), False Positives (FP), False Negatives (FN)
TP = np.diag(conf_matrix)
FP = np.sum(conf_matrix, axis=0) - TP
FN = np.sum(conf_matrix, axis=1) - TP
TN = np.sum(conf_matrix) - (TP + FP + FN)

# Accuracy
accuracy = np.sum(TP) / np.sum(conf_matrix)

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# F1 Score
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
import numpy as np

# Confusion matrix
conf_matrix = np.array([[1787, 980, 614], [960, 2279, 142], [634, 122, 2625]])

# Chỉ số cho từng lớp
for i in range(conf_matrix.shape[0]):
    TP = conf_matrix[i, i]
    FP = np.sum(conf_matrix[:, i]) - TP
    FN = np.sum(conf_matrix[i, :]) - TP
    TN = np.sum(conf_matrix) - (TP + FP + FN)
    
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1_score = 2 * (precision * recall) / (precision + recall)
    
    print("Class:", i+1)
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1_score)
    print()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['B2', 'B5', 'B6']
accuracy_labels = [0.69, 0.78, 0.85]
f1_score_labels = [0.53, 0.67, 0.78]
overall_accuracy = 0.66

# Plot
fig, ax = plt.subplots(figsize=(12, 5))

# Plot bars
x = np.arange(len(labels))
width = 0.35
ax.bar(x - width/2, accuracy_labels, width, label='Accuracy', color='skyblue')
ax.bar(x + width/2, f1_score_labels, width, label='F1 Score', color='orange')

# Plot horizontal line for overall accuracy
ax.axhline(overall_accuracy, color='red', linestyle='--', label='Overall Accuracy')

# Annotate bars with values
for i, (accuracy, f1_score) in enumerate(zip(accuracy_labels, f1_score_labels)):
    ax.text(i - width/2, accuracy + 0.01, str(round(accuracy, 2)), ha='center', va='bottom', color='black', fontweight='bold')
    ax.text(i + width/2, f1_score + 0.01, str(round(f1_score, 2)), ha='center', va='bottom', color='black', fontweight='bold')

# Set labels and legend
# ax.set_xlabel('Labels')
# ax.set_ylabel('Scores')
ax.set_title('Accuracy và F1 Score cho mỗi nhãn', fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels)
legend = ax.legend(loc='upper left')
plt.setp(legend.texts, fontweight='bold')

# Set y-axis limit
ax.set_ylim(0.5, 0.9)

# Annotate overall accuracy
overall_text = ax.text(len(labels)-2, overall_accuracy - 0.03, f'Overall Accuracy: {overall_accuracy}', ha='center', va='bottom', color='black')
plt.setp(overall_text, fontweight='bold')
# Show plot
plt.show()
```
